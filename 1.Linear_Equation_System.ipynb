# System of Linear Equations

The values of $x$ and $y$ in the equation $2x + 3y = 28$ are the same as in $x + 2y = 17$, $5$ y $6$. Together, these equations can make a system of linear equations.

The corresponding system of equations is shown below.

\begin{cases}
2x + 3y = 28 \\
x + 2y = 17
\end{cases}

FYI: $2x + 3y = 28$ is equivalent to $y = -2/3x + 28/3$. In this sense, what's an equivalent form of $x + 2y = 17$?

### 1. Graph of a system of linear equations
The graph of the above system of linear equations is shown below. 

```python
#
# Non-interactive.
#
x = range(11)
y_1 = lin_rec(-2/3, 28/3, *x)
y_2 = lin_rec(-1/2, 17/2, *x)

y_3 = y_1 + y_2
plt_2lin_rec(0, min(x), max(x), min(y_3), max(y_3), x, y_1, y_2, 
             lbl_1='2x+3y=28', lbl_2='x+2y=17', intrx=5, intry=6)
```

The point $(5, 6)$ corresponds to the intersection of the lines $2x + 3y = 28$ and $x + 2y = 17$. In other words, the solution for the system of linear equations $2x + 3y = 28$ and $x + 2y = 17$ is the pair of $(x,y)$ coordinates $(5, 6)$.

### 1.1. Parallel lines
The pair of linear equations, $y = x/2 - 2$ and $y = x/2 + 1$, are a pair of parallel lines. For these lines, their slope is the same ($m = 1/2$) but their y-intercept is different from each other ($-2$ and $1$, respectively). The graph for this pair of parallel lines is shown below.

```python
#
# Non- interactive. 
#
x = range(11)
y_1 = lin_rec(.5, -2, *x)
y_2 = lin_rec(.5, 1, *x)

y_3 = y_1 + y_2
plt_2lin_rec(0, min(x), max(x), min(y_3), max(y_3), x, y_1, y_2, 
             lbl_1='y=x/2-2', lbl_2='y=x/2+1', intrx=5, intry=6)
```

### 1.2. The same line
The pair of linear equations, $y - x/2 = -2$ and $4y - 2x = -8$ are in fact the same line. The equation $4y = 2x - 8$ results from multiplying the equation $y = x/2 - 2$ by $4$. The graph of these equations is shown below. 

```python
#
# Non-interactive.
#
x = list(range(11))
y = lin_rec(0.5, -2, *x)

plt_lin_rec(min(x), max(x), min(y), max(y), x, y, m=0.5, b=-2)
```
